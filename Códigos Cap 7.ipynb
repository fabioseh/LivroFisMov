```python
from math import *
import matplotlib.pyplot as plt
import numpy as np
```

```python
m0=5000
b=100
vr=1000


def v(t):
    return vr*log(m0/(m(t)))

def m(t):
    return m0-b*t


T=np.arange(0,40,1e-3)

V=[v(t) for t in T]
M=[m(t) for t in T]

fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.plot(T,V, color=color)
ax1.set_xlabel("Tempo (s)", color="black")
ax1.set_ylabel("Velocidade (m/s)", color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx() 
color = 'tab:blue'
ax2.plot(T,M, color=color)
ax2.set_xlabel("Tempo (s)", color="black")
ax2.set_ylabel("Massa (kg)", color=color)
ax2.tick_params(axis='y', labelcolor=color)

plt.grid()
plt.show()
```
